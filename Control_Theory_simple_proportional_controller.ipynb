```python
import numpy as np
import matplotlib.pyplot as plt

# Define system parameters
m = 1  # Mass of object
b = 1  # Damping coefficient
k = 1  # Spring constant
F = 1  # Applied force

# Define controller gains
kp = 1

# Define simulation parameters
dt = 0.01  # Time step
t = np.arange(0, 10, dt)  # Time vector
x = np.zeros(len(t))  # Position vector
v = np.zeros(len(t))  # Velocity vector
x[0] = 1  # Initial position
v[0] = 0  # Initial velocity

# Simulate system with proportional controller
for i in range(len(t)-1):
    xdot = v[i]
    vdot = (F - b*v[i] - k*x[i] - kp*x[i])/m
    x[i+1] = x[i] + v[i]*dt
    v[i+1] = v[i] + vdot*dt

# Plot results
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.show()
```
